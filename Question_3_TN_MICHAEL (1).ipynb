```python
import matplotlib.pyplot as plt
data = np.linspace(-5,5,10)
cov = cov_1(data)
_, ax = plt.subplots(1, 2, figsize=(20, 20))
ax = np.ravel(ax)
ax[0].plot(data, np.zeros_like(data), 'ko')
ax[0].set_yticks([])
for idx, i in enumerate(data):
    ax[0].text(i, 0+0.005, idx)
ax[0].set_xticks(data)
#ax[0].set_xticklabels(data)
ax[0].set_xticklabels(np.round(data, 2), rotation=70)

ax[1].grid(False)
im = ax[1].imshow(cov)
colors = ['w', 'k']
for i in range(len(cov)):
  for j in range(len(cov)):
      ax[1].text(j, i, round(cov[i, j], 2),
      color=colors[int(im.norm(cov[i, j]) > 0.5)],
      ha='center', va='center', fontdict={'size': 16})
ax[1].set_xticks(range(len(data)))
ax[1].set_yticks(range(len(data)))
ax[1].xaxis.tick_top()
```

The first kernel produces a symmetrical heatmap, with the diagonal values taking the larger values. This is due to each element in the covariance matrix being inversely proportional to the distance between the two points, and the diagonal being the result of comparing each data point with itself. 
This particular kernel is appropriate for modelling smooth functions. 
Remember that the kernel is translating the distance of the data points along the x axis of covariances for values on y axis.
Thus, the closer two points on the x axis, the more similar we expect them to be on the y-axis. We can see then that data points 3 to 9 are extremely far apart, with a covariance of 0. Thus when generating the data, they could even be seen as outliers to the rest of the data.

# Question 3.3

```python
x_1 = [-4.0,-2.0,4.0]
y = [-0.5,0.5,0.0]
```

Kernel function

```python
n = 100
#prior
x = np.linspace(-4,4,n)

#Kernel
def Kern(X):
  return np.array([np.exp(-(X-x_i)**2/2) for x_i in X])
```

```python
k_p = Kern(x)
mu_p = np.zeros_like(x)
y_p = np.random.multivariate_normal(mu_p, k_p, 10)
```

```python
X_joint = np.r_[x_1, x]
K_joint = Kern(X_joint)
```

```python
n1 = len(x_1)
K_j = K_joint[:n1, :n1]
K_x = K_joint[n1:, :n1]
K_xx = K_joint[n1:, n1:]
```

Posteior mean

```python
mu_post = K_x.dot(np.linalg.inv(K_j)).dot(y)
```

Posterior covariance matrix

```python
k_post = K_xx - K_x.dot(np.linalg.inv(K_j)).dot(K_x.T)
```

Plotting posterior mean with data points to show Gaussian process has recovered mean function

```python
import matplotlib.patches as mpatches
plt.plot(x, mu_post, color='blue')
plt.scatter(x_1,y, color='red')
obs = mpatches.Patch(color='red', label = 'Observed Values')
m_p = mpatches.Patch(color='blue', label = 'Posterior mean')
plt.legend(handles=[obs, m_p], loc = 'upper right')
plt.title('Posterior mean vs Observed data points')
plt.show()
```

Plotting posterior mean with 1 and 2 standard deviations for quantification of uncertainty

```python
std = np.sqrt(np.diag(k_post))
```

```python
_,ax = plt.subplots(2,1, figsize=(10,10))
ax[0].fill_between(x, mu_post + std, mu_post - std, color='orange');
ax[0].plot(x, mu_post, color='black')
std1 = mpatches.Patch(color='orange', label = 'Posterior mean with 1 std')
ax[0].legend(handles=[std1], loc='upper right');
ax[1].fill_between(x, mu_post + 2*std, mu_post-2*std, color='red');
ax[1].plot(x, mu_post, color='black')
std2 = mpatches.Patch(color='red', label = 'Posterior mean with 2 std')
plt.legend(handles=[std2], loc='upper right');
plt.title('Posterior mean with uncertainty measurement')
plt.show()
```
